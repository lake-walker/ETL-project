```python
# Dependecies 
# import dependencies and modules
from matplotlib import pyplot as plt
import numpy as np
import scipy.stats as stats
import pandas as pd
import time
import requests
from pprint import pprint
```

```python
# Import happiness data from csv
happy_2015 = pd.read_csv('data/happy_data/2015.csv')
happy_2016 = pd.read_csv('data/happy_data/2016.csv')
happy_2017 = pd.read_csv('data/happy_data/2017.csv')
happy_2018 = pd.read_csv('data/happy_data/2018.csv')
happy_2019 = pd.read_csv('data/happy_data/2019.csv')

# Rename country or regioin column into country 
happy_2015 = happy_2015.rename(columns={'Country': 'country'})
happy_2016 = happy_2016.rename(columns={'Country': 'country'})
happy_2016.dtypes
```

```python
# Import suicide data from csv
full_sui_data = pd.read_csv('data/suicide_data.csv')

# Create two new dataframes to hold the 2015 and 2016 data
sui_2015 = full_sui_data[full_sui_data['year'] == 2015]
sui_2016 =  full_sui_data[full_sui_data['year'] == 2016]

# Join the suicide data with the happiness data
sui_2016
```

```python
# Join the the suicide data and happiness data
data_2015 = sui_2015.merge(happy_2015, on='country')
data_2016 = sui_2016.merge(happy_2016, on='country')
data_2015
```

```python
for col in data_2016.columns:
    print(col)
```

```python
# Drop useless columns
data_2015 = data_2015.drop(columns=['country-year','HDI for year'])
data_2016 = data_2016.drop(columns=['country-year','HDI for year'])
```

```python
# clean_2015_data = data_2015.dropna()
data_2016.dropna()
data_2015
```

```python
for col in data_2015.columns:
    print(col)
```

```python
# Concat the 2015 and the 2016 data
full_data = pd.concat([data_2015, data_2016])
full_data.dtypes
```

```python
# Create new dataframes in a relational format in order to move to SQL and a relational database
# Country id
country_df = full_data['country']
country_df = pd.DataFrame(country_df)
# Drop all the duplicates to get one piece of country data
clean_country = country_df.drop_duplicates()
# Generate ids for each country 
ids = np.random.randint(low=0, high=100, size = len(clean_country['country']))
clean_country['country_id'] = ids
clean_country.head()
```

```python
# Create ids for year
year_df = full_data['year']
year_df = pd.DataFrame(year_df)
# Drop duplicates
clean_years = year_df.drop_duplicates()
# Generate ids
year_ids = np.random.randint(low=1000, high=9999, size = len(clean_years['year']))
clean_years['year_id'] = year_ids
clean_years
```
